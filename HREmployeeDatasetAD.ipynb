```python
'''
HR Employee Dataset

Dataset: https://www.kaggle.com/dushantkumar/hrdataset-v13

When I first started with this dataset, the plan was to use it for both a Tableau visual and 
for other analytical assessments. For some reason, I decided to not use it for analytical 
assessments, but I kept the code for that anyways.
'''


import pandas as pd
import numpy as np

data = pd.read_csv('~/Desktop/Sklearn Data Sets/data analytics data sets/HR Employee Data/HRDataset.csv')
#print("This is the initial unchanged data file: \n", data.head())

#Remove unnecessary columns
#data=data.iloc[:, np.r_[0:12,15:16,18:24,27,28,31:33]]
print(data.info())
print(data.head())
```

```python
uniqueColumns = data.iloc[:, np.r_[1:26,28:34]]
for col in uniqueColumns:
    uniqueValues= data[col].unique().tolist()
    length = len(uniqueValues)
    print("The number of unique values in the",col, "column is",length,". Those values are:",uniqueValues,'.\n')
```

```python
#convert CitizenDesc from string to integer
#Create a dictionary for the changes
citizenNumber={'US Citizen':1, 'Eligible NonCitizen':2, 'Non-Citizen':3}
data['CitizenDesc'].replace(citizenNumber, inplace=True)

#convert HispanicLatino values to all lowercase
YesNo= {'yes':1, 'Yes':1, 'no': 0, 'No':0}
data['HispanicLatino'].replace(YesNo, inplace=True)

#convert ManagerID values from float to int
data['ManagerID'] = data['ManagerID'].convert_dtypes(convert_integer=True)
print(data)
```

```python
#Convert Recruitment Source to an integer
data['RecruitmentSource'] = data['RecruitmentSource'].astype('category')
data.dtypes
data['RecruitmentSourceNumber'] = data['RecruitmentSource'].cat.codes
data.head()
```

```python
#Convert RaceDesc to an integer
#The number of unique values in the RaceDesc column is 6 . Those values are: ['Black or African American', 
#'White', 'Asian', 'American Indian or Alaska Native', 'Two or more races', 'Hispanic'] .

data['RaceDesc'] = data['RaceDesc'].astype('category')
data.dtypes
data['RaceDescNumber'] = data['RaceDesc'].cat.codes
data.head()
```

```python
#Convert Term Reason to an integer
data['TermReason'] = data['TermReason'].astype('category')
data.dtypes
data['TermReasonNumber']=data['TermReason'].cat.codes
data.head()
```

```python
#Convert State to a integers
data['State'] = data['State'].astype('category')
data.dtypes
data['StateNumber']=data['State'].cat.codes
data[['State','StateNumber']].head()
```

```python
#trim leading and trailing spaces from employee names
data['Employee_Name'] = data['Employee_Name'].str.strip()
print(data)
```

```python
#Convert state to a numeric value
data['State'] = data['State'].astype('category')
data['StateNumber']=data['State'].cat.codes
print(data)
```
